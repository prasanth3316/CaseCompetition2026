# Load the Transactions data 

```python
tblTransactions = pd.read_csv("../exported_data/tblTransactions.csv")
tblTransactions = tblTransactions.fillna(0)
tblTransactions.head()
```

```python
tblTransactions["major_brand"] = tblTransactions["major_brand"].astype(str)
```

```python
fig, ax = plt.subplots(4,1, figsize=(12,10))
ax[0].plot(tblTransactions["modality"], tblTransactions["candy_retail_spend"])
ax[0].set_title("Total candy retail spend per modality")
ax[0].set_ylabel("candy retail spend")

ax[1].plot(tblTransactions["modality"], tblTransactions["chicken_retail_spend"])
ax[1].set_title("Total chicken retail spend per modality")
ax[1].set_ylabel("chicken retail spend")

ax[2].plot(tblTransactions["modality"], tblTransactions["cheese_retail_spend"])
ax[2].set_title("Total cheese retail spend per modality")
ax[2].set_ylabel("cheese retail spend")

ax[3].plot(tblTransactions["modality"], tblTransactions["frzn_chicken_retail_spend"])
ax[3].set_title("Total frozen chicken retail spend per modality")
ax[3].set_ylabel("frozen chicken retail spend")

plt.savefig("../exported_charts/comparison-modality-each-items")

plt.tight_layout()

plt.show()
```

# Analysis 
- For candy retail spending
-> We can observer costumers tend to spend more through delivery

- For chicken retail spending
-> We can observe costumers tend to spend more in instore

- For cheese retail spending 
-> We can observe costumers tend to spend in similar amounts 

- For frozen chicken retail spending 
-> We can ovbserve costumers tend to spend slightly more instore followed by pickup and delivery
